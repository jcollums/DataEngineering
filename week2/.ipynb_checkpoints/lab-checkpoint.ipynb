```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "http://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
# Print out the text
text = soup.get_text()
print(soup.text)
```

```python
soup.find_all('a')
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
# Print the first 10 rows for sanity check
rows = soup.find_all('tr')
print(rows[:10])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
row_results = []
for row in rows:
    tds = row.find_all('td')
    
    cells = []
    for td in tds:
        text = td.getText().strip()
        cells.append(text)
    row_results.append(cells)
print(row_results)
```

```python
df = pd.DataFrame(row_results)
df.head(10)
```

```python
headers = soup.find_all("th")
print(headers)
```

```python
h_results = []
for header in headers:
    cell = header.getText().strip()
    h_results.append(cell)
    
print(h_results)
```

```python
frames = [pd.DataFrame([h_results]), pd.DataFrame(row_results)]
df4 = pd.concat(frames)
df4.head(10)
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
df5.info()
```

```python
df5.shape
```

```python
df7 = df5.drop(df5.index[0])
df7.head()
```

```python
from datetime import datetime
```

```python
df8 = df7.dropna(axis=0, how='any')
df8.info()
df8.head()
```

```python
time_list = df8['Chip Time'].tolist()
time_mins = []
for i in time_list:
    hms = i.split(':')
    h = 0
    if len(hms) == 3:
        h = hms[0]
        m = hms[1]
        s = hms[2]
    else:
        m = hms[0]
        s = hms[1]
    math = (int(h) * 3600 + int(m) * 60 + int(s))/60
    time_mins.append(math)
print(time_mins)
```

```python
df8['Runner_mins'] = time_mins
df8.head()
```

```python
df8.describe(include=[np.number])
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 15, 5
df8.boxplot(column='Runner_mins')
plt.grid(True, axis='y')
plt.ylabel('Chip Time')
plt.xticks([1], ['Runners'])
```

```python
x = df8['Runner_mins']
ax = sns.distplot(x, hist=True, kde=True, rug=False, color='m', bins=25, hist_kws={'edgecolor':'black'})
plt.show()
```

```python
f_fuko = df8.loc[df8['Gender']=='F']['Runner_mins']
m_fuko = df8.loc[df8['Gender']=='M']['Runner_mins']
sns.distplot(f_fuko, hist=True, kde=True, rug=False, hist_kws={'edgecolor':'black'}, label='Female')
sns.distplot(m_fuko, hist=False, kde=True, rug=False, hist_kws={'edgecolor':'black'}, label='Male')
plt.legend()
```

```python
g_stats = df8.groupby("Gender", as_index=True).describe()
print(g_stats)
```

```python
df8.boxplot(column='Runner_mins', by='Gender')
plt.ylabel('Chip Time')
plt.suptitle("")
```
